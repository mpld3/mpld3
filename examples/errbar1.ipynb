```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# example data
x = np.arange(0.1, 4, 0.5)
y = np.exp(-x)
```

```python
error = 0.1 + 0.2 * x
```

```python
fig, ax = plt.subplots()
ax.errorbar(x,y,yerr=error,fmt="^")
plt.show()
```
